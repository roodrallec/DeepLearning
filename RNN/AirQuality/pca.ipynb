```python
import numpy as np
import pandas as pd
from scipy.io import arff
from scipy import linalg 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
raw_data = np.load('LondonAQ.npz')
sites = raw_data.keys()
df = pd.DataFrame()
_vars = ['PM10', 'NO', 'NO2', 'O3', 'Wind Speed']

for site in sites:
    for var in _vars:
        column = site + '_' + var
        var_data = raw_data[site][:, _vars.index(var)]
        df[column] = var_data
        
df.head()
```

```python
X = df[["GEltham_PM10", "GWesthorne_PM10"]].values
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_ratio_)
```

```python

for site in sites:
    for var in _vars:
        site_var = site + "_" + var
        df["GEltham_PM10"].corr(df['B'], method='')
        pca.fit(df[[, site_var]].values)
        print(site_var, pca.explained_variance_ratio_)
```
